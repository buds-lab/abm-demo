```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from common.utils import load_variable
from energy_models import Zone
```

# Load experiment results

# Unmet Comfort Ratio (UNC)


![](img/image.png)

Introduced in [LightLearn: An adaptive and occupant centered controller for lighting based on reinforcement learning](https://www.sciencedirect.com/science/article/abs/pii/S0360132318306462?via%3Dihub)

The **LOWER** the **BETTER**

## Baseline (tolernace 0.35, stochastic timings)

```python
zones, dict_zones_unc, dict_zones_unc_avg = {}, {}, {}
exp = 'baseline_tolerance_week'
zones[exp] = load_variable(exp + "/" + "Zone_1.pkl")
dict_zones_unc[exp] = load_variable(exp + "/" + "dict_zones_unc.pkl")
dict_zones_unc_avg[exp] = load_variable(exp + "/" + "dict_zones_unc_avg.pkl")


cols = dict_zones_unc["baseline_tolerance_week"].keys()
df = pd.DataFrame(columns=cols)

# unc per occupant
df_aux = pd.DataFrame(dict_zones_unc[exp], index=[exp])
df = df.append(df_aux)
    
print(df)

fig, ax = plt.subplots(figsize=(20,6))
sns.heatmap(df, vmin=0, vmax=1, annot=True, ax=ax)
plt.xlabel("Occupants")
plt.ylabel("Escenarios")
plt.savefig(f"img/heatmp_unc_{exp}")
```
